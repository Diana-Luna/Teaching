## 5.3 Introduction to Matplotlib

We start by importing the libraries we use in this lesson.

````python
import numpy as np
import matplotlib.pyplot as plt
````

In our first example we will plot the function $y=x^2$ for the values $x=0,1,2,3,4,5$.

We prepare our data by defining two lists:

````python
x=[0,1,2,3,4,5]  #Independent variable
y=[0,1,5,9,16,25] #Dependent variable
````

We now plot as follows:

````python
plt.plot(x,y) #plt.plot(x,y) takes lists/arrays x and y and automatically plots the data
plt.xlabel('x') # label the x-axis with the string provided
plt.ylabel('f(x)=x^2') # label the y-axis wiht the string provided.
plt.show() # show the plot
````

Let us understand how the previous lines of code work. First, `plt.plot(x,y)` reads and automatically plots the data. Functions `plt.xlabel()` and `plt.ylabel()` each take one string as input and use it to label the $x$- and $y$-axis, respectively. Finally, `plt.show()` shows the plot in our screen. 

Plotting in Matplotlib is like preparing a cake: *We can add as many layers as we like*. For instance,  we can add a graph with the cubes of the same $x$-values as follows:

````python
z=[0,1,8,27,64,125]
plt.plot(x,y) # Plot each (x,y) value
plt.plot(x,z) # Plot each (x,z) value
plt.xlabel('x')
plt.ylabel('y=x^2 and z=x^3')
plt.show()
````

Matplotlib automatically adjusts the axes to fit our graphs. Also, by default, `plt.plot()` generates a *line graph* connecting contiguous $(x,y)$-values. This can be adjusted by providing a third string argument to the `plt.plot` encoding our desired layout:

````python
plt.plot(x,y,'ro') # r=Red, o=markers are points.
plt.plot(x,z,'gx') # g=Green, x=markers are xs.
plt.xlabel('x') 
plt.ylabel('y=x^2 and x=x^3')
plt.show()
````

The third string argument for the `plt.plot` function encodes the display as folloes:

- First character sets the color. For instance, `'r'` stands for *red* and `'g'` for *green*.
- Second character sets the color. For instance, `'o'` stands for *point plot* and `'x'` for using *crosses*.

We refer to a full list of colors and markers styles we refer to 

https://github.com/matplotlib/cheatsheets/raw/master/cheatsheets-1.png

It is important to keep in mind that Matplotlib provides different methods for building the same plot. For instance, the next method is more readable at the cost of being lengthier. 

````python
plt.plot(x,y,color='red',marker='o',linestyle='')
plt.plot(x,z,color='green',marker='x',linestyle='')
plt.xlabel('x')
plt.ylabel('y=x^2 and x=x^3')
plt.show()
````

**Quiz:** Can you figure out why the previous plot need the linestyle argument?

When we plot multiple functions it is good practice to add a lengend to identify them. This is done by providing a `label` argument to `plt.plot()` and by adding the `plt.legend()` layer.

````python
plt.plot(x,y,color='red',marker='o',linestyle='',label='x^2')
plt.plot(x,z,color='green',marker='x',linestyle='',label='x^3')
plt.xlabel('x')
plt.ylabel('y=x^2 and x=x^3')
plt.legend()
plt.show()
````

## 5.3  1D array creation in NumPy


Enter five data points manually is easy. However, in real life we either generate data using the computer or we obtain data from an external source. In this section we will see how we can use NumPy to create data. Afterwards we will see how to import data from a file.

### 5.3.1 Introduction to `np.array`s

NumPy has the `ndarray` datatype which is very similar to a Python list. We will call them *arrays* for simplicity.

````python
nparray=np.array([1,2,3.1416,4,5])
````

NumPy arrays are very similar to Python list. One can for instance read a particular entry and slice a sublist by using the same syntax as for Python lists:

````python
print(nparray[2])
print(nparray[2:])
````

We can also assign new values.

````python
nparray[2]=3
print(nparray)
````

We can even loop through their elements.

````python
for element in array:
    print(element)
````

However, not all that works for lists works for arrays. For instance, we cannot concatenate arrays having entries of different types. 

````python
# Python method for concatenating two lists
a=[1,2,3]
b=['four','five','six']
a+b
````

````python
# Numpy method for concatenating two arrays
a=np.array([1,2,3])
b=np.array(['four','five','six'])
np.concatenate(a,b)
````

Indeed, even if you try the same Python method with arrays of the same type, you may obtain an unexpected result:

````python
a=np.array([1,2,3])
b=np.array([4,5,6])
a+b
````

Ok, so *why should we use arrays instead of lists?* Simply because NumPy arrays are more efficient for performing numerical computations. But as Uncle Ben from Spider-Man said, *with great power comes great responsibility*, that is, NumPy numeric power comes at the cost of having less flexibility than for lists. ### 5.3.2 Be careful with types

As seen before, arrays require a careful approach when dealing with types inside the array. Each array has an assigned type that can be seen using the `.dtype` attribute:


````python
int_array=np.array([0,1,2,3,4])
print(int_array.dtype)
````

````python
float_array=np.array([3.1416, 1.4142,2.7182])
print(float_array.dtype)
````

````python
boolean_arr=np.array([True,False, False, True])
print(boolean_arr.dtype)
````

Types `int64`, `float64`, and `bool` are integer, float and boolean, respectively, and we will restrict to these types for this course. 
Other types can be found in https://numpy.org/devdocs/user/basics.types.html. 
 Note that NumPy automatically guesses which type the array should be. One can also specify the type by adding the `dtype` argument to the `np.array` method:

````python
floaty=np.array([0,1,2,3,4],dtype='float64')
print(floaty.dtype)
````

### 5.3.3 Arithmetic and 1D arrays

One advantage of using NumPy arrays instead of list is that we can perform arithmetic operations in their elements. For instance, the following methods can be used for adding, multiplying, and dividing an array by an scalar:

````python
a=np.array([1,2,3])
print(a+10)
````

````python
print(2*a)
````

````python
print(a/2)
````

*Element-wise* operations are also possible:

````python
a=np.array([1,2,3])
b=np.array([4,5,6])
````

````python
#element-wise sum
print(a+b)
````

````python
#element-wise product (also known as dot product)
print(a*b)
````

Even rising all the entries  to a given power:

````python
a**2
````

### 5.3.4 Creating 1D arrays from NumPy functions

Here we enlist a few handy NumPy methods for creating arrays:

#### Integer arrays

 **`np.zeros(length,dtype)`:** Generates an array full of zeroes of length `length` and type `dtype` (last argument is optional). 

````python
print(np.zeros(10) )
````

`np.ones(length,dtype)`: Same as before but fills with ones instead.

````python
print(np.ones(5))
````

**Quiz**: Generate a `np.array` with a hundred `2`s.

**`np.arange`:** Similar to Python's `range` function,  `np.arange(n)` creates an array of numbers from `0` to `n-1`.

````python
np.arange(10)
````

Also `start_value`,  `stop_value` and `step_size` can be specified as:
```python
np.arange(start_value,stop_value,step_size)
```

````python
a=np.arange(5,15,2)
print(a)
````

**`np.random.randint(a, b, size)`:** returns random integers from the interval $[a,b)$. The `size` argument indicates the length of the output array.

````python
np.random.randint(3,10,8)
````

#### Float arrays

**`np.arange`:** This method can also be used to produce float arrays. For instance

````python
float_array=np.arange(0.,1.,0.1)
print(float_array)
````

generates an array with numbers from `0` to `1`, where the difference between two consecutive values is `0.1`.

Indeed, `np.arange` can be used to draw graphs of mathematical functions. For instance, suppose that we are plotting $y=x^2$ and $z=x^3$ in the interval $[-1,1]$. For this we subdivide  $[-1,1]$ into a fine array of points as follows:

````python
x=np.arange(-1.,1.,0.01)
````

Then, we compute the values of $y$ and $z$ on these points:

````python
y=x**2
z=x**3
````

Finally, we plot:

````python
plt.plot(x,y,label='y=x^2')
plt.plot(x,z,label='z=x^3')
plt.xlabel('x')
plt.legend()
plt.show()
````

**`np.linspace(start,stop,num)`:** This method gives an alternative way to subdivide an interval. It produces a list of evenly spaced floats from  `start` to `float`. The parameter `size`  specifies the number of points considered.

For instance,  to plot  $y=sin(x)$ in the interval $[-4\pi,4\pi]$, we generate 100 evenly spaced points in that interval as follows.

````python
x=np.linspace(-4*np.pi, 4*np.pi,100)
````

Then, compute $sin(x)$ in each point.

````python
y=np.sin(x)
````

We note that `np.sin(x)` reads as computing $\sin(\cdot)$ for each value inside the vector $x$. This is called a *vectorized*, and we will learn about them in Section XX. 

Now we plot.

````python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y=sin(x)')
plt.show()
````

**`np.random.random(size)`:** Draws floats from the interval $[0,1)$ uniformly at random. The output is an array of length `size`.

````python
np.random.random(10)
````

**`np.random.normal(mean,sd,size)`:** Draws floats from a normal (Gaussian) distribution with mean `mean` and standard deviation `sd`. 

````python
a=np.random.normal(0,1,100)
print(a)
````

We can indeed "verify" that our data comes from a normal distribution by plotting a histogram using the `hist` function:

````python
plt.hist(a)
plt.show()
````
